```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
con=PorterStemmer()
```

```python
def conver(wor):
    y=[]
    for i in wor.split():
        y.append(con.stem(i))
    stor=" ".join(y)
    return stor
```

```python
conver("dacned")
```

```python
#conver(mov['tags'])
```

```python
mov['tags']=mov['tags'].apply(conver)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
tex=CountVectorizer(max_features=450,stop_words='english')
```

```python
arr=tex.fit_transform(mov['tags']).toarray()
arr
```

```python
tex.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import  cosine_similarity
```

```python
cos=cosine_similarity(arr)
```

```python
cos[0],arr
```

```python

def rec(movie):
    
    movie_ind=mov[mov['Title'] == movie].index[0]
    
    dsit=cos[movie_ind]
    
    #dsit=dsit.astype(np.int64).all()
    lis=sorted(list(enumerate(dsit)),reverse=True,key=lambda x:x[1])[1:6]
    for i in lis: 
        print(mov.iloc[i[0]].Title)
            
```

```python

rec('Avengers: Infinity War')
	
	

```
